## Examine 2nd order derivative & DBSCAN

```python
import os, glob, getpass, sys
import warnings
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table, join, vstack, hstack, Column, MaskedColumn, unique
from astropy.utils.exceptions import AstropyWarning

user = getpass.getuser()
sys.path.append('/Users/' + user + '/Dropbox/my_python_packages')
path = '../../'
from gaia.cluster_dbscan import Dbscan as Dbscan
```

```python
def set_plot(fontsize = 26, xlim = 0, ylim = 1, xlabel = 'xaxis', ylabel = 'xaxis', xstep = None, ystep = None, 
             legend = None, title = None):
    plt.ylim(ylim)
    plt.xlim(xlim)
    plt.yticks(fontsize = fontsize)
    plt.xticks(fontsize = fontsize)
    plt.xlabel(xlabel, fontsize = fontsize)
    plt.ylabel(ylabel, fontsize = fontsize)
    if title:
        plt.title(title, fontsize = fontsize) 
    if xstep:
        plt.xticks(np.arange(min(xlim), max(xlim)+1, xstep))
    if ystep:
        plt.yticks(np.arange(min(ylim)+0.01, max(ylim)+0.01, ystep))
    if legend:
        plt.legend(loc = 'upper right', fontsize = fontsize*0.9, markerscale = 1.5, handletextpad = 0.1, numpoints = 1)
```

```python
# Read data ====================================
warnings.simplefilter('ignore', AstropyWarning)
data     = Table.read(path  + 'sample_gaia/gaia_sample_cleaned.vot',       format = 'votable')
data_scl = Table.read(path  + 'sample_clusters/cl_scale/scl_standard.vot', format = 'votable')

data_x   =  Dbscan()
data_x.load(data=data, data_scl=data_scl.to_pandas())
```

```python
# After trial-and-error =========================
# See Notebook 02_test_eps
dbscan_pars_0 = {'mpts':100, 'bin_step':2, 'bin_gauss':2, 'peak_thres':0.5,  'ylim_1':[0.32, 0.45]}
dbscan_pars_1 = {'mpts':130, 'bin_step':1, 'bin_gauss':2, 'peak_thres':0.5, 'ylim_1':[0.36, 0.47]}
dbscan_pars_2 = {'mpts':160, 'bin_step':1, 'bin_gauss':2, 'peak_thres':0.5, 'ylim_1':[0.40, 0.50]}

dbscan_pars = dbscan_pars_1
mpts        = dbscan_pars['mpts']
bin_step    = dbscan_pars['bin_step']
bin_gauss   = dbscan_pars['bin_gauss']
peak_thres  = dbscan_pars['peak_thres']
ylim_1      = dbscan_pars['ylim_1']
```

```python
# PLOT ==========================================
data_inp = data_x
fig      = plt.figure(figsize=[25,9])
ftsize   = 36
colors   = iter(['blue', 'limegreen', 'red'])


# First Plot ================================
ax = plt.subplot(121)
plt.subplots_adjust(wspace = 0.3)

for mpts in [100, 130, 160]:
    data_inp.compute_neighbours(min_points = mpts)
    x     = data_inp.n_distance['ind'].data[0::5]  #rebin every 5 steps to ease PDF plots
    y     = data_inp.n_distance['dist'].data[0::5]
    label = r'$k$ = ' + np.str(mpts)
    plt.plot(x, y, 'o', label = label, markeredgecolor = 'none', markersize = 8, color = next(colors))

set_plot(xlim=[1000, len(data_inp.data)], ylim=[0.25, 1.2], fontsize = ftsize, xlabel = 'Point',
         ylabel = r'$k$-Distance', title = 'Gaia Sample', xstep = 250, legend = True)
plt.grid()
ax.axvspan(2000, 2130, alpha=0.25, color='grey')


# Second Plot ===============================
plt.subplot(122)
data_inp.compute_neighbours(min_points=dbscan_pars['mpts'])
data_inp.kdist_tb_rebin(bin_step=dbscan_pars['bin_step'])
data_inp.kdist_tb_smooth(gauss=dbscan_pars['bin_gauss'])
data_inp.deriva_2()

peaks   = data_x.get_deriva2_peaks(plot=False, xrg=[2000, 2150], thres=dbscan_pars['peak_thres'], min_dist=7)

plt.plot(data_inp.opoints, data_inp.kdist, 'o', color = 'limegreen', label = r'$k$ = ' + np.str(mpts), markersize = 10, markeredgecolor = 'green')
set_plot(xlim=[2000, 2130], ylim=[0.35,0.50], fontsize = ftsize, xlabel = 'Point', ylabel = r'$k$-Distance',
         title = 'Zoom-in', xstep = 30, ystep = 0.04, legend = True)

i = 2
peak_x = peaks[i]
peak_y = data_x.kdist[data_x.opoints == peaks[i]]

plt.axvline(x=peak_x, ymin=0, ymax=2,    linestyle = '--', color = 'black')
plt.axhline(y=peak_y, xmin=0, xmax=2200, linestyle = '--', color = 'black')

plt.show()
fig.savefig('03_plot_eps_0.pdf', bbox_inches = 'tight', overwrite = True)
```
