## Examine 2nd order derivative & DBSCAN - Final, paper result

```python
import os, glob, getpass, sys
import warnings
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table, join, vstack, hstack, Column, MaskedColumn, unique
from astropy.utils.exceptions import AstropyWarning

user = getpass.getuser()
sys.path.append('/Users/' + user + '/Dropbox/my_python_packages')
from gaia.cluster_dbscan import Dbscan as Dbscan
```

```python
# Read data ====================================
warnings.simplefilter('ignore', AstropyWarning)
path     = '/Users/' + user + '/Dropbox/Public/clustering_oph/'
data     = Table.read(path  + 'sample_gaia/gaia_sample_cleaned.vot',       format = 'votable')
data_scl = Table.read(path  + 'sample_clusters/cl_scale/scl_standard.vot', format = 'votable')
```

```python
# After trial-and-error =========================
dbscan_pars_0 = {'mpts':100, 'bin_step':2, 'bin_gauss':2, 'peak_thres':0.5, 'ylim_1':[0.32, 0.45], 'min_dist':7}
dbscan_pars_1 = {'mpts':130, 'bin_step':1, 'bin_gauss':2, 'peak_thres':0.5, 'ylim_1':[0.36, 0.47], 'min_dist':7}
dbscan_pars_2 = {'mpts':160, 'bin_step':1, 'bin_gauss':2, 'peak_thres':0.5, 'ylim_1':[0.40, 0.50], 'min_dist':7}
```

```python
# Compute 2nd derivative and plot ===============
dbscan_pars = dbscan_pars_0

mpts        = dbscan_pars['mpts']
bin_step    = dbscan_pars['bin_step']
bin_gauss   = dbscan_pars['bin_gauss']
peak_thres  = dbscan_pars['peak_thres']
ylim_1      = dbscan_pars['ylim_1']
min_dist    = dbscan_pars['min_dist']
xlim        = [2000, 2150]
ylim_2      = [0.01,0.4]


data_x   = Dbscan()
data_x.load(data=data, data_scl=data_scl.to_pandas())

data_x.compute_neighbours(min_points=mpts)
data_x.kdist_tb_rebin(bin_step=bin_step)
data_x.kdist_tb_smooth(gauss=bin_gauss)
data_x.deriva_2()

peaks = data_x.get_deriva2_peaks(plot=False, xrg=xlim, thres=peak_thres, min_dist=7)


# PLOT ==========================================
fig    = plt.figure(figsize=[20,7])
ftsize = 26

plt.subplot(121)
plt.subplots_adjust(wspace = 0.3)
plt.xlim(xlim)
plt.ylim(ylim_1)
plt.xlabel('Point', fontsize = ftsize)
plt.ylabel(r'$k$-Distance', fontsize = ftsize)
plt.xticks(fontsize = ftsize)
plt.yticks(fontsize = ftsize)
plt.xticks(np.arange(min(xlim), max(xlim)+1, 50.0))

plt.plot(data_x.opoints[0::bin_step], data_x.kdist[0::bin_step], 'o', color = 'darkorange', label = r'$k$ = ' + np.str(mpts))
plt.legend(loc = 'upper right', fontsize = ftsize*0.9, markerscale = 1.5, handletextpad = 0.1, numpoints = 1)
for peak in peaks:
    plt.axvline(x=peak, ymin=0, ymax=2, linestyle = '--', color = 'black')
# =============================================
plt.subplot(122)
plt.xlabel('Point', fontsize = ftsize)
plt.ylabel(r'$k$-Distance $2^{nd}$ derivative', fontsize = ftsize)
plt.xlim(xlim)
plt.ylim(ylim_2)

plt.xticks(fontsize = ftsize)
plt.yticks(fontsize = ftsize)
plt.xticks(np.arange(min(xlim), max(xlim)+1, 150.0))

plt.plot(data_x.opoints, data_x.kdist_d2, linewidth = 3)

for peak in peaks:
    plt.axvline(x=peak, ymin=0, ymax=2, linestyle = '--', color = 'black')
    
plt.show()

for peak in peaks:
    print(f'Point = {peak} ; EPS = {data_x.kdist[data_x.opoints == peak][0]:>6.3f}')
    
# fig.savefig('benchmark_3_0.pdf', bbox_inches = 'tight', overwrite = True)
```

```python
# Run DBSCAN ====================================
eps   = np.float(input('Introduce EPS: '))
db_test = Dbscan()
db_test.load(data = data, data_scl = data_x.data_scl)
db_test.add_label()
db_test.run_dbscan(eps=eps, mpts=mpts)
db_test.get_clusters()
```
