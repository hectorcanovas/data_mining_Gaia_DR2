## Apply HDBSCAN

```python
import os, glob, getpass, sys
import warnings
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table, join, vstack, hstack, Column, MaskedColumn, unique
from astropy.utils.exceptions import AstropyWarning

user = getpass.getuser()
sys.path.append('/Users/' + user + '/Dropbox/my_python_packages')
from gaia.cluster_eps import EPS
path = '../../'
```

# Read and scale the data

```python
# Read real data ===============================
warnings.simplefilter('ignore', AstropyWarning)
features = ['X', 'Y', 'Z', 'pmdec', 'pmra']

# Scale data with Standard Scaler ==============
for scaler in ['standard', 'minmax', 'robust']:
    data_scl =  EPS()
    data_scl.read_data_tb(path + 'sample_gaia/gaia_sample_cleaned.vot')
    data_scl.select_features(features)
    data_scl.scl_data(scaler_mode = scaler, verbose = True)
    data_scl.plot_distributions(pdf_name='scl_' + scaler + '.pdf')
    data_scl.write_scaled(filename='scl_' + scaler + '.vot')
```
