## CMD Diagram + Isochrones +  Stellar Tracks

```python
import os, glob, getpass, sys, warnings

import itertools
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table, join, vstack, hstack, Column, MaskedColumn, unique
from astropy.utils.exceptions import AstropyWarning
from astropy import units as u
from astropy.visualization import hist

user = getpass.getuser()
sys.path.append('/Users/' + user + '/Dropbox/my_python_packages')
path = '../'
sys.path.append(path)
from gaia.stellar import Stellar as Stellar
from gaia.Basic import Basic as Basic
from photometry.Extinction import Extinction as Ext
from extra_codes import sample_comp    as samp_comp
```

```python
# Path to data =================================
warnings.simplefilter('ignore', AstropyWarning)
path_0        = path + 'sample_control/'
path_1        = path + 'sample_clusters/cl_'
path_2        = path + 'sample_gaia/'

path_control  = path_0 + 'OPH___control_sample.vot'
path_gaia     = path_2 + 'gaia_sample_cleaned.vot'
path_entire   = 'tab_3.vot'

# Read Data ====================================
sample_gaia    = Basic()
sample_gaia.read_gaia_cat(path_gaia)

sample_control = Table.read(path_control, format = 'votable')
sample_entire  = Table.read(path_entire,  format = 'votable')
sample_common  = sample_entire[sample_entire['DOH'] == 'YYY']
```

```python
# Examine Photometric Errors ==================
fig      = plt.figure(figsize=[30,7])
ftsize   = 18
ymax     = 1400
inp_cols = ['phot_g_mean_mag_err', 'phot_bp_mean_mag_err', 'phot_rp_mean_mag_err']

for i in range(3):
    ax      = plt.subplot(131 + i)
    inp_col = inp_cols[i]
    samp_comp.plot_1_hist_07(sample_gaia.cat, axis_obj=ax, inp_col=inp_col)
    print('Mean Error in ' + inp_col + ': ', '{:>5.4f}'.format(sample_gaia.cat[inp_col].mean()))
    
plt.show()
print('============= MEAN ERROR (DASHED LINE) IS > THAN HISTOGRAM PEAK =============')
```

```python
# Combine with Gaia Cat to get all cols ========
sample_control = join(sample_gaia.cat, Table([sample_control['source_id']])) ; sample_control.label = 'Control'
sample_common  = join(sample_gaia.cat, Table([sample_common['source_id']]))  ; sample_common.label  = 'Common'
sample_entire  = join(sample_gaia.cat, Table([sample_entire['source_id']]))  ; sample_entire.label  = 'Combined'
```

```python
# Read Isochrones ==============================
ages = [2,5]
isos = []
for age in ages:
    isoc = Stellar()
    isoc.read_bt_iso(age=age)
    isoc.read_pc_iso(age=age, mass_above = 1.4)
    isoc.combine_iso()
    isos.append(isoc.iso_comb)

# Construct Stellar Tracks =====================
track_list = []
masses     = [0.1, 0.5, 1.4]
for mass in masses:
    star_tracks = Stellar()
    star_tracks.track_mass(mass=mass)
    track_list.append(star_tracks.stellar_track)
```

```python
# Compute Extinction vector =====================
test   = Ext()
Av     = 5.0 # See notebook #06
Ag     = Av * test.filters['A_l_A_v_f99_I05'][test.filters['Phot_nm'] == 'Gmag'][0]
print(f'Computing Extinction for Av = {Av} (Ag = {Ag:.1f})')


test.redden_mag(Av=Av, phot_nm='Gmag',   phot_mag=10)
test.redden_mag(Av=Av, phot_nm='Grmag',  phot_mag=10)

# Vector Coordinates ==========
x_0 = test.Gmag   - test.Grmag     # Original
x_1 = test.Gmag_e - test.Grmag_e   # Observed
y_0 = test.Gmag                    # Original
y_1 = test.Gmag_e                  # Observed

dx       = 0.01
dy       = 0.0
xy       = (x_1 + dx, y_1 + dy)
xytext   = (x_0 + dx, y_0 + dy)
```

```python
# Extract Control NOT in Common ================
sample_control_nc       = join(sample_control, sample_common['ra', 'source_id'], keys = 'source_id', join_type='left')
sample_control_nc       = sample_control_nc[sample_control_nc['ra_2'].mask == True]
sample_control_nc.label = 'Control'
sample_control_nc.remove_column('ra_2')
sample_control_nc.rename_column('ra_1', 'ra')
print(f'N Control Sources NOT in common: {len(sample_control_nc)}')

clusters    = [sample_gaia.cat, sample_control_nc, sample_common]
```

```python
# Plot Color-Magnitude diagram =================
fig        = plt.figure(figsize=[7,7])
col_1      = 'phot_g_mean_mag_abs'
col_2      = 'phot_bp_mean_mag_abs'
col_3      = 'phot_rp_mean_mag_abs'
m_sizes    = [4, 2, 2] # 5 is larger than photometric error (see above)
fontsize   = 18
linewidth  = 3
arrowprops = {'facecolor':'black', 'width':2, 'headwidth':12, 'headlength':14}


ax = plt.subplot(111)
samp_comp.plot_stl_07(xlabel = r'$G - G_\mathrm{RP}$ [mag]', ylabel = r'$M_\mathrm{G}$ [mag]', fontsize = fontsize, ylim = [16,0], xlim = [-0.1,2.7])

# Plot Data =======================
plt.plot(clusters[0][col_1]   - clusters[0][col_3],   clusters[0][col_1], 'o', color = 'black',   alpha = 0.2, label = 'Gaia',    markersize = m_sizes[0])
plt.plot(clusters[1][col_1]   - clusters[1][col_3],   clusters[1][col_1], 'o', color = 'magenta', alpha = 1.0, label = 'Control', markersize = m_sizes[1])
plt.plot(clusters[2][col_1]   - clusters[2][col_3],   clusters[2][col_1], 'o', color = 'cyan',    alpha = 1.0, label = 'Common',  markersize = m_sizes[2])

# Plot Isochrones =================
colors     = itertools.cycle(['blue', 'blue'])
linestyles = itertools.cycle(['-', '--'])
for iso in isos:
    plt.plot((iso[col_1] - iso[col_3]), iso[col_1], label = iso.label, linewidth = 2, linestyle = next(linestyles), color = next(colors))

# Plot Evolutionary tracks ========
color = 'darkorange' #'darkorange' # 'limegreen'
for track in track_list:
    plt.plot(track[col_1] - track[col_3], track[col_1], linestyle = '-', linewidth = 4, color = color, label = '')

plt.legend(loc = 'upper right', fontsize = fontsize, handlelength = 1.0)


# Plot Extinction vector =====
ax.annotate('', xy = xy, xytext = xytext,arrowprops = arrowprops)
dx = -0.05
dy = -0.3
ax.annotate(r'$A_\mathrm{G} =$' + f'{Ag:4.1f}', xy=xy, xytext=(xytext[0] + dx, xytext[1] + dy), fontsize = fontsize)

# Annotate evolutionary tracks 
weight = 'bold'
xy_1   = (0,0)
ax.annotate(np.str(masses[2]) + ' M' + np.str(r'$_\mathbf{\odot}$'),xy = xy_1, xytext=(0.7, 2.8), fontsize = fontsize, color = color, weight = weight)
ax.annotate(np.str(masses[1]) + ' M' + np.str(r'$_\mathbf{\odot}$'),xy = xy_1, xytext=(0.9, 4.9), fontsize = fontsize, color = color, weight = weight)
ax.annotate(np.str(masses[0]) + ' M' + np.str(r'$_\mathbf{\odot}$'),xy = xy_1, xytext=(1.4, 8.5), fontsize = fontsize, color = color, weight = weight)

plt.show()
fig.savefig('07_create_CMD_0.pdf', bbox_inches = 'tight', overwrite = True)
```

```python
# To estimate Objects mass below 0.5 Msun evolutionary track
els = sample_common['phot_g_mean_mag'] - sample_common['phot_rp_mean_mag'] > 1.05 # Approx location of 0.Msun track
print(f'Objects below the 0.5Msun evolutionary track: {len(sample_common[els]) / len(sample_common) * 100:5.1f}%')
```

# Comparison with previous studies 

```python
cats     = [samp_comp.read_tb_07(path + 'sample_initial/'+ inp) for inp in ['2005_wilking_Table4.vot', '2011_erickson_Table3.vot']]
splots   = iter([121,122])
fig      = plt.figure(figsize =[30,5])
fontsize = 18

for cat in cats:
    inp_m  = cat['Mass'].mean()
    inp_s  = cat['Mass'].std()
    inp_pc = len(cat[cat['Mass'] < 0.5])/len(cat) * 100

    print(f'{cat.label}: Average Stellar Mass: {inp_m:14.1f} +/- {inp_s:3.1f}')    
    print(f'{cat.label}: Fraction of objects < 0.5Msun: {inp_pc:6.1f}%')
    print()

    ax       = plt.subplot(next(splots))
    histplot = hist(cat['Mass'], 'knuth', edgecolor='black')
    plt.title(cat.label, fontsize = fontsize)
    plt.xticks(fontsize = fontsize)
    plt.yticks(fontsize = fontsize)

    ax.axvline(x=0.5, color = 'k', linestyle = '--')

plt.show()
```
