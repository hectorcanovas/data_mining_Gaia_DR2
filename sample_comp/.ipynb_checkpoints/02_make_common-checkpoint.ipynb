## Construct common sample

```python
import os, glob, getpass, sys
import warnings
import numpy as np
import matplotlib.pyplot as plt
import itertools
from astropy.table import Table, join, vstack, hstack, Column, MaskedColumn, unique
from astropy.utils.exceptions import AstropyWarning

user = getpass.getuser()
sys.path.append('/Users/' + user + '/Dropbox/my_python_packages')
path = '../'
sys.path.append(path)
from gaia.cluster_comparison_n import Comparison_n as Comparison
from extra_codes import sample_comp as samp_comp
```

```python
# Path to data =================================
warnings.simplefilter('ignore', AstropyWarning)
path_0    = path + 'sample_control/'
path_1    = path + 'sample_clusters/cl_'
cols      = ['ra', 'dec', 'parallax', 'pmra', 'pmdec', 'source_id']

path_control  = path_0 + 'OPH___control_sample.vot'
path_wise_img = path_0 + 'wise_RGB_img.fits'


# Control Data =================================
path_control  = path_0 + 'OPH___control_sample.vot'


# Cluster Data =================================
tabs_dbscan   = glob.glob(path_1 + 'dbscan/cl_dbscan*vot')   ; tabs_dbscan.sort()
tabs_hdbscan  = glob.glob(path_1 + 'hdbscan/cl_hdbscan*vot') ; tabs_hdbscan.sort()
tabs_optics   = glob.glob(path_1 + 'optics/cl_optics*vot')   ; tabs_optics.sort()
```

```python
# Read Data ====================================
label_db = ['DBSCAN;  mPts = ' + inp[inp.rfind('_')+1:inp.find('.vot')] for inp in tabs_dbscan]
label_op = ['OPTICS;  mPts = ' + inp[inp.rfind('_')+1:inp.find('.vot')] for inp in tabs_optics]
label_hd = ['HDBSCAN; mCls = ' + inp[inp.rfind('_')+1:inp.find('.vot')] for inp in tabs_hdbscan]

    
cl_dbscan_l, cl_optics_l, cl_hdbscan_l  = [],[],[]
for i in range(len(tabs_dbscan)):
    cl_dbscan_l.append (samp_comp.read_cluster(inp_path = tabs_dbscan[i],  label = label_db[i], cols=cols))
    cl_optics_l.append (samp_comp.read_cluster(inp_path = tabs_optics[i],  label = label_op[i], cols=cols))
    cl_hdbscan_l.append(samp_comp.read_cluster(inp_path = tabs_hdbscan[i], label = label_hd[i], cols=cols))

control = samp_comp.read_cluster(inp_path = path_control, label = 'Control', cols=cols)



# Select clusters according to the case ========
col_0 = [len(inp) for inp in cl_dbscan_l]
col_1 = [len(inp) for inp in cl_optics_l]
col_2 = [len(inp) for inp in cl_hdbscan_l]

table_els = Table(data=[col_0, col_1, col_2], names=['DBSCAN', 'OPTICS', 'HDBSCAN'])
indexes_0 = [np.where(col == np.min(col))[0][0]                          for col in [col_0, col_1, col_2]] # Min.    Case
indexes_1 = [np.where((col != np.max(col)) & (col != np.min(col)))[0][0] for col in [col_0, col_1, col_2]] # Average Case
indexes_2 = [np.where(col == np.max(col))[0][0]                          for col in [col_0, col_1, col_2]] # Max.    Case
table_els
```

```python
# Read individual tables for paper =============

tab_d = Table.read(path_1 + 'dbscan/output_dbscan.tex',   format = 'ascii.latex')
tab_o = Table.read(path_1 + 'optics/output_optics.tex',   format = 'ascii.latex')
tab_h = Table.read(path_1 + 'hdbscan/output_hdbscan.tex', format = 'ascii.latex')

tab_h['Eps'] = [0] * len(tab_h)
tab_h.rename_column('mCls', 'mPts')

tab_cls = vstack([tab_d, tab_o, tab_h])
tab_cls['Control_pc'] = ['{:2.1f}'.format(inp/len(control)*100.) for inp in tab_cls['Control']]
tab_cls
```

```python
# Select case to combine following Table 2 in the paper
case_r = np.str(input('introduce cluster combination (min/mean/max): '))
if case_r != 'max' and case_r != 'mean' and case_r != 'min':
    print('Wrong Option. Try again')
if case_r == 'min':
    case = indexes_0 + ['_case_0']
if case_r == 'mean':
    case = indexes_1 + ['_case_1']
if case_r == 'max':
    case = indexes_2 + ['_case_2']

print('Combining Clusters with ' + case_r + ' amount of elements')
cluster_list = [cl_dbscan_l[case[0]], cl_optics_l[case[1]], cl_hdbscan_l[case[2]]]
```

```python
# Compare Outputs ==============================
clusters = Comparison()
clusters.load_clusters(control=control, cluster_list = cluster_list)
clusters.read_wise_fits(path_wise_img)


# Create Plots =================================
clusters.plot_wise_img_3(markersize=10, mew=3, mec_1 = 'yellow', mec_2='lightgrey', figname='02_make_common_0' + case[3] + '.pdf', ftsize=28, control_on_top = False, legend_color='limegreen')

colors = itertools.cycle(["blue", "limegreen", "red"])
hists  = clusters.plot_3_hist(x1_bins=5, x2_bins=5, x3_bins=5, fig_nm='02_make_common_1' + case[3] + '.pdf', x2lim=[-20,0],
                     x3lim=[-32.5,-17.5], linewidth=5, colors=colors, ftsize = 38)
```

## Create Combined Sample ===============

```python
# Create Combined & Common Sample ==============
clusters.combine_all() # Creates clusters.all    table
clusters.get_common()  # Creates clusters.common table
clusters.get_uncommon()
clusters.get_new()
```

```python
# Create Cluster-Detection label ===============
entire       = clusters.all.copy()
cluster_list = [inp            for inp in clusters.clusters] + [clusters.control]
labels       = [inp.label[0:1] for inp in cluster_list[0:3]] + [clusters.control.label]
labels       = iter(labels)

for cluster in cluster_list:
    label          = next(labels)
    cluster[label] = ['Y'] * len(cluster)
    
    entire = join(entire, cluster['source_id', label], keys='source_id', join_type='left')
    entire[label][entire[label].mask == True] = 'N'
    

# Merge labels =================================
entire['DOH']    = [entire['D'][i] + entire['O'][i] + entire['H'][i] for i in range(len(entire))]
entire.remove_columns(['D','O','H'])

for col in ['ra', 'dec', 'parallax', 'pmra', 'pmdec']:
    entire[col].format = '10.7f'  # To fit table in 1 pdf page

entire.sort('ra')
entire[0:3]
```

```python
# Merge labels =================================
print(f'Entire Sample:         {len(entire):>16.0f} Elements')
print(f'Control in Entire:     {len(entire[entire["Control"] == "Y"]):16.0f} Elements')
print(f'Entire - Only DBSCAN:  {len(entire[entire["DOH"] == "YNN"]):16.0f} Elements')
print(f'Entire - Only OPTICS:  {len(entire[entire["DOH"] == "NYN"]):16.0f} Elements')
print(f'Entire - Only HDBSCAN: {len(entire[entire["DOH"] == "NNY"]):16.0f} Elements')
      
common = entire[entire['DOH'] == 'YYY']
print()
print(f'Common:                {len(common):16.0f} Elements')
print(f'Control in Common:     {len(common[common["Control"] == "Y"]):16.0f} Elements')
print(f'New??:                 {len(common[common["Control"] == "N"]):16.0f} Elements')
```

```python
# Save Tables ==================================
filename = 'entire_sample' + case[3]
entire.write(     filename + '.vot', format = 'votable',     overwrite = True)
entire[0:5].write(filename + '.tex', format = 'ascii.latex', overwrite = True)
print('Saving Entire sample as ' + filename + '.tex/.vot')


filename = 'common_sample' + case[3]
common.write(     filename + '.vot', format = 'votable',     overwrite = True)
common[0:5].write(filename + '.tex', format = 'ascii.latex', overwrite = True)
print('Saving Common sample as ' + filename + '.tex/.vot')
```

```python
# Prepare table to search for "new" targets in SIMBAD ====
new           = common[common['Control'] == 'N']
filename      = filename + '_no_control_simbad.txt'
new['Simbad'] = ['Gaia dr2 ' + np.str(inp) for inp in new['source_id']]

Table([new['Simbad']], names=['Simbad ID']).write(filename, format = 'ascii.fast_no_header',overwrite = True)
print('Saving NEW sample as '    + filename)
print()
print('THIS TABLE IS CROSSED AGAINST SIMBAD TO FIND TARGETS WITH NO SIMBAD ENTRIES (NEW ONES)')
print()
print('THE CROSS*SIMBAD OUTPUT IS SAVED IN THIS DIRECTORY AS simbad.xml')
```
