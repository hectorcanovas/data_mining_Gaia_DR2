## Compile Photometry & Disc Classification ==========

```python
import os, glob, getpass, sys, warnings

import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table, join, vstack, hstack, Column, MaskedColumn, unique
from astropy.utils.exceptions import AstropyWarning
from astropy import units as u

user = getpass.getuser()
sys.path.append('/Users/' + user + '/Dropbox/my_python_packages')
path = '../'
sys.path.append(path)

from gaia.photometry import Photometry as phot
from extra_codes     import sample_comp as samp_comp
```

```python
# Path to data =================================
warnings.simplefilter('ignore', AstropyWarning)
path_wise   = path + '../extra_data/WISE_data/WISE_ALL/'
path_gaia   = path + 'sample_gaia/gaia_sample_cleaned.vot'
path_tab_ir = path + 'sample_comp/08_IR_crossmatch.vot'           # IR Photometry Entire Table
path_tab_SN = path + 'sample_comp/09_make_WISE_maps_0_w4_SN.vot'  # IR WISE-4 SN  Table


# Read Data ====================================
sample_gaia  = Table.read(path_gaia)
common_ir    = Table.read(path_tab_ir) # IR photometry table
common_SN    = Table.read(path_tab_SN) # IR photometry table

# Apply SN criteria ============================
sn_limit  = 4
strs      = 14
print('=' * strs + ' WARNING ' + '=' * strs)
print(f'Selecting Sources with WISE W4 SN > {sn_limit}')
common_SN = common_SN[common_SN['SN_W4'] > sn_limit]
print('=' * strs + ' WARNING ' + '=' * strs)
print()

print(f'N Targets with 2MASS & WISE photometry BEFORE SN selection: {len(common_ir):7.0f}')
common_ir = join(common_SN['source_id', 'artifact', 'SN_W4'], common_ir, keys='source_id')

print(f'N Targets with 2MASS & WISE photometry AFTER SN selection: {len(common_ir):8.0f}')
print(f'Control targets: {len(common_ir[common_ir["control"] == "Y"]):50.0f}')
```

```python
# Show WISE 4 IMAGES ===========================
table_wise = common_SN
band       = 'w4'
filename   = '10_make_photometry_0_' + band + '.pdf'
wise_phot  = samp_comp.wise_multiphot_09(path_wise=path_wise, table_wise=table_wise, band=band, filename=filename,
                                         zoom_size=1.5* u.arcmin, figsize=[30,35], nrows=7, ncols=7)
wise_phot.rename_column('SN', 'SN_W4')
```

```python
# Merge Gaia & IR-Photometry tables ============
cols_1 = ['source_id', 'phot_g_mean_mag', 'phot_g_mean_flux_over_error',
        'phot_bp_mean_mag', 'phot_bp_mean_flux_over_error', 'phot_rp_mean_mag', 'phot_rp_mean_flux_over_error']

common_ir = join(sample_gaia[cols_1], common_ir, keys='source_id')

# Add discovery flag ===========================
common_ir['status'] = ['New    '] * len(common_ir)

for i in range(len(common_ir)):
    if common_ir['control'][i] == 'Y': common_ir['status'][i] = 'Control'
    if (common_ir['control'][i] == 'N') and (common_ir['NB_REF'][i] != 0): common_ir['status'][i] = 'Other'

common_ir.write('10_make_photometry_tb_0.vot', format = 'votable', overwrite = True)
common_ir[0:3]
```

```python
# Create Photometry Table ======================
gaia_phot = phot()
gaia_phot.read_catalogue('10_make_photometry_tb_0.vot')
gaia_phot.compute_gaia_emag()
gaia_phot.compute_wise_ccor()
gaia_phot.mag_2_flux()
gaia_phot.apply_wise_ccor()
gaia_phot.apply_2mass_sys_err()
gaia_phot.apply_wise_sys_err()
gaia_phot.add_wavefreq()
gaia_phot.add_lFlambda()
gaia_phot.compute_alpha_w34()
gaia_phot.compute_alpha_w14()
gaia_phot.compute_alpha_kw4()
gaia_phot.remove_cols(['f_c(W1)', 'f_c(W2)', 'f_c(W3)', 'f_c(W4)']) # Flux Correction Cols
gaia_phot.cat[0:3]
```

```python
# Identify Disc Classes ========================
phot_tb   = gaia_phot.cat
alpha_IR  = 'alpha_kw4' # Traditional alpha slope, from 2 - 22 micrometers
class_III = phot_tb[alpha_IR]  < -1.6
class__II = (phot_tb[alpha_IR] > -1.6)  & (phot_tb[alpha_IR] <-0.3)
class__FS = (phot_tb[alpha_IR] >= -0.3) & (phot_tb[alpha_IR] <= +0.3)
class___I = (phot_tb[alpha_IR] >= 0.3)

phot_tb['DClass_0'] = ['            '] * len(phot_tb)
phot_tb['DClass_0'][class_III] = 'class_III'
phot_tb['DClass_0'][class__II] = 'class__II'
phot_tb['DClass_0'][class__FS] = 'class__FS'
phot_tb['DClass_0'][class___I] = 'class___I'

print(f'Number of class_III: {len(phot_tb[class_III]):10.0f}')
print(f'Number of class__II: {len(phot_tb[class__II]):10.0f}')
print(f'Number of class__FS: {len(phot_tb[class__FS]):10.0f}')
print(f'Number of class___I: {len(phot_tb[class___I]):10.0f}')
print()
```

```python
# Classification refinement ====================
alpha_IR  = 'alpha_w34'       # alpha slope from WISE3 to WISE4 [12-22 micrometers]
group_0   = phot_tb[alpha_IR]  < -1.6                              # Bare Photosphere
group_1   = (phot_tb[alpha_IR] > -1.6) & (phot_tb[alpha_IR] <0.3)  # Discs
group_2   = phot_tb[alpha_IR]  >= 0.3                              # Disc with Au-sized cavity

phot_tb['DClass_1'] = ['Photosphere'] * len(phot_tb)
phot_tb['DClass_1'][group_0] = 'Photosphere'
phot_tb['DClass_1'][group_1] = 'Disc'
phot_tb['DClass_1'][group_2] = 'Cavity'
```

```python
# Adopt Final Classification ===================
phot_tb['SED_Class'] = phot_tb['DClass_0']
phot_tb['SED_Class'][phot_tb['DClass_1'] == 'Photosphere'] = 'Photosphere'
phot_tb['SED_Class'][phot_tb['SED_Class'] == 'class_III'] = 'Class III'
phot_tb['SED_Class'][phot_tb['SED_Class'] == 'class__II'] = 'Class  II'

els_0     = phot_tb['SED_Class'] == 'Photosphere'
els_1     = phot_tb['SED_Class'] == 'Class III'
els_2     = phot_tb['SED_Class'] == 'Class  II'


# Save entire photometry table =================
phot_tb.write('10_make_photometry_tb_1.vot', format = 'votable', overwrite = True)

print()
print(f'Number of BarePhots: {len(phot_tb[els_0]):10.0f}')
print(f'Number of Class III: {len(phot_tb[els_1]):10.0f}')
print(f'Number of Class  II: {len(phot_tb[els_2]):10.0f}')

print()
print('NOTE: FOR SIMPLICITY I AM MERGING TRANSITIONAL DISCS WITH CLASS II')
```

```python
# Print WISE MAG limits ========================
print()
print('======== Min/Max [Ks] - [W4] colors ======')
class_str = iter(['class_III', 'class_II '])
for inp in [class_III, class__II]:
    sub_tb  = phot_tb[inp]
    color_m = np.min(sub_tb['Ksmag'] - sub_tb['W4mag'])
    color_M = np.max(sub_tb['Ksmag'] - sub_tb['W4mag'])
    dclass  = next(class_str) 
    print(f'{dclass} {color_m:5.1f} {color_M:5.1f}')

print()
print('======== Min/Max [W3] - [W4] colors ======')
class_str = iter(['group_0', 'group_1', 'group_2'])
for inp in [group_0, group_1, group_2]:
    sub_tb = phot_tb[inp]
    color_m = np.min(sub_tb['W3mag'] - sub_tb['W4mag'])
    color_M = np.max(sub_tb['W3mag'] - sub_tb['W4mag'])
    gclass  = next(class_str) 
    print(f'{gclass} {color_m:5.1f} {color_M:5.1f}')
```

```python
# Make magnitude - color plot ==================
fig       = plt.figure(figsize=[7,7])
fontsize  = 18
ax        = plt.subplot(111)

plt.ylabel('[Ks] - [W4]', fontsize = fontsize)
plt.xlabel('[W3] - [W4]', fontsize = fontsize)
plt.tick_params(axis='both', which='major', labelsize=fontsize)


# Plot data ====================================
filename = '10_make_photometry_1.pdf'
cels     = phot_tb['control'] == 'Y' # Control Elements
col_x    = phot_tb['W3mag'] - phot_tb['W4mag']
col_y    = phot_tb['Ksmag'] - phot_tb['W4mag']
plt.ylim([0,6.5])

plt.plot(col_x,       col_y,       'o', markersize = 5,  color = 'cyan',     label = 'Common')
plt.plot(col_x[cels], col_y[cels], 'o', markersize = 3,  color = 'magenta',  label = 'Control')

lim_x1 = 1.00
lim_x2 = 2.40 # Real value is 2.45 (see above) but the plot looks clearer with 2.40

ax.axhline(y=3.1,   xmin=-10, xmax=+10, linestyle = ':',  color = 'k')
ax.vlines(x=lim_x1, ymin = 0,  ymax=3.1, linestyle = '--', color = 'k')

ax.annotate("Class  II",       xy=(0,0), xycoords='data',xytext=(0.0,  6.0), fontsize = fontsize)
ax.annotate("Class III",       xy=(0,0), xycoords='data',xytext=(3.0,  2.5), fontsize = fontsize)
ax.annotate("Photospheres",    xy=(0,0), xycoords='data',xytext=(-0.07,2.5), fontsize = fontsize*0.8)

plt.legend(loc = 'lower right', fontsize = fontsize, handletextpad = -0.3)
plt.show()
fig.savefig(filename, bbox_inches = 'tight', overwrite = True)
```
