# Plot NIR + ODISEA TARGETS

```python
# For Sky Plots se: http://docs.astropy.org/en/stable/visualization/wcsaxes/ticks_labels_grid.html
import numpy as np
from astropy.table import Table, join, vstack, hstack, Column, MaskedColumn
import warnings
from astropy.utils.exceptions import AstropyWarning
import os, glob, getpass, sys
from astroquery.skyview import SkyView
from astropy import units as u
from astropy.io import fits
import matplotlib.pyplot as plt
from astropy.visualization import MinMaxInterval, SqrtStretch,ImageNormalize, ManualInterval, PercentileInterval

user = getpass.getuser()
sys.path.append('/Users/' + user + '/Dropbox/my_python_packages')
from gaia.Basic_Plotters import Basic_Plotters
path      = '../'
path_wise = '/Users/' + user + '/Astrofisica/papers/2019_clustering_oph/extra_data/WISE_data/wise_OPH_map/'
```

```python
# Read Reference Sample ========================
warnings.simplefilter('ignore', category=AstropyWarning)
tb_control = Table.read(path + 'sample_control/OPH___control_sample.vot', format = 'votable') ; tb_control.label = 'Control'
```

```python
# Use SkyView to download NIR images: https://skyview.gsfc.nasa.gov/current/cgi/query.pl
# See also: http://docs.astropy.org/en/stable/visualization/wcsaxes/overlays.html
#See also: http://astroweb.case.edu/jakub/TA/Query_databases.html
download = False
if download:
    labels  = ['WISE 3.4', 'WISE 4.6', 'WISE 12']
    ndir    = 'wise_OPH_map'
    os.system('mkdir ' + ndir)

    for label in labels:
        img = SkyView.get_images("247.0 -24.0", survey=label, radius=7*u.degree)
        img[0][0].writeto(ndir + '/' + np.str(label.replace(' ', '_')) + '.fits')
```

```python
# Read Images ==================================
text = 'NOTE THIS DATA IS STORED IN A SEPARATE DIRECTORY AS IT IS TOO HAVE TO BE UPDALOADED TO GITGUB'
print('=' * len(text))
print(text)
print('=' * len(text))

files  = glob.glob(path_wise + '*')
fits_l = [fits.open(inp) for inp in files]
imgs   = [np.nan_to_num(img[0].data) for img in fits_l]

# Find best strecht ============================
fig    = plt.figure(figsize=[30,5])
xlims  = [[0,+70], [0,+70], [750,+1100]]
ftsize = 18
for i in range(len(imgs)):
    plt.subplot(131+i)
    plt.hist(imgs[i].flatten(), 800)
    plt.xlabel('count value', fontsize = ftsize)
    plt.ylabel('N Counts', fontsize = ftsize)
    plt.xticks(fontsize = ftsize)
    plt.yticks(fontsize = ftsize)
    plt.xlim(xlims[i])

plt.show()
```

```python
# Re-scale Images ==============================
vmins     = [15, 20, 820]
vmaxs     = [80,75,1400]
imgs_scl  = []
imgs_sqrt = []


for i in range(len(imgs)):
    interval = ManualInterval(vmin=vmins[i], vmax=vmaxs[i]) # Normalize [0:1] the image within the given threshold
    stretch  = SqrtStretch()+ PercentileInterval(98.)       # Apply sqrt Strecht

    imgs_scl.append(interval(imgs[i]))
    imgs_sqrt.append(stretch(interval(imgs[i])))

# Create & Plot final images ===================
imgs_scl.append(( np.dstack((imgs_scl[2], imgs_scl[1], imgs_scl[0])) *255).astype(np.uint8)) # Linear Strecht
imgs_sqrt.append((np.dstack((imgs_sqrt[2],imgs_sqrt[1],imgs_sqrt[0]))*255).astype(np.uint8)) # Sqrt   Strecht
    
fig = plt.figure(figsize=[30,15])
for i in range(len(imgs_scl)):
    plt.subplot(241 + i)
    plt.imshow(imgs_scl[i], origin='lower')

    plt.subplot(241 + i + 4)
    plt.imshow(imgs_sqrt[i], origin='lower')

plt.show()
```

```python
# Plot final Image =============================    
figs  = Basic_Plotters(tb_control)
figs.load_wise_img(wise_img=imgs_sqrt[3], wise_head=fits_l[0][0].header)
figs.plot_wise_img(mec='yellow', mew=2, markersize=7, show_gradii=True, figname='02_plot_odisea_img_0.pdf', ftsize=16)

# Write fits image =============================
hdu    = fits.PrimaryHDU(data=imgs_sqrt[3], header=fits_l[0][0].header)
hdu.writeto('wise_RGB_img.fits', overwrite=True)
```
