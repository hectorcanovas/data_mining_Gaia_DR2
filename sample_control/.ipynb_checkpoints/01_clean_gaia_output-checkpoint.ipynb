# Read Gaia Cross Match & Clean Samples

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy import units as u
from astropy.table import Table, join, vstack, hstack, Column, MaskedColumn, unique
import warnings
from astropy.utils.exceptions import AstropyWarning
from astropy.visualization import hist
import getpass, sys
user = getpass.getuser()
sys.path.append('/Users/' + user + '/Dropbox/my_python_packages')
path = '../'
from gaia.Basic import Basic
from gaia.Basic_Plotters import Basic_Plotters
```

```python
# Read Reference Sample ========================
warnings.simplefilter('ignore', category=AstropyWarning)
path_ini = path + 'sample_initial/sample_initial_oph_x_gaia-result.vot'
initial  = Table.read(path_ini, format = 'votable')

# Read Gaia-DR2 * with Reference Sample ========
control = Basic()
control.read_gaia_cat(path_ini)

print(f'Total INPUT  DR2 Targets: {len(initial):27.0f}')
print(f'Total OUTPUT DR2 Targets: {len(control.cat):27.0f}')
print(f'Missing Targets in DR2:   {len(initial) - len(control.cat):27.0f}')

# Remove duplicates ============================
control.remove_duplicates(key='source_id')
# control.remove_duplicates(key='col2mass')
```

```python
# Plot J-band magnitude distribution =========== 
merged   = join(initial['ra', 'dec', 'col2mass', 'jmag'], control.cat, keys='col2mass', join_type='outer')
merged.rename_column('jmag_1', 'jmag')
merged_1 = merged[merged['source_id'].mask == True]  # NOT In Gaia
merged_2 = merged[merged['source_id'].mask == False] # In Gaia


# Plot histogram of J-band distribution ==========================
fig = plt.figure(figsize=[10,5])
ftsize = 22
plt.xticks(fontsize = ftsize)
plt.yticks(fontsize = ftsize)
plt.xlabel(r'$m_\mathrm{J}$ [magnitudes]', fontsize = ftsize)
plt.ylabel('# Sources', fontsize = ftsize)

plt.hist(merged_2['jmag'], label = 'in Gaia DR2',    histtype = 'step', linewidth = 3)
plt.hist(merged_1['jmag'], label = 'not in Gaia DR2', histtype = 'step', linewidth = 3)
plt.legend(loc = 'upper left', fontsize = ftsize*0.75)
plt.show()
fig.savefig('01_clean_gaia_output_0.pdf', overwrite = True, bbox_inches = 'tight')
```

```python
# Compute Quality controls =====================
control.get_BPRP2()
control.get_unit_weight_error()
control.get_umax_criteria()
control.get_ruwe_0()

# Apply 1st quality-control ====================
control.apply_criteria_visper(minvis=7) # Precise criteria
control.get_len()
control.apply_PSN_cut(PSN=10)        # Reliable criteria
control.get_len()

# Apply 2nd quality-control ====================
control.apply_criteria_ruwe()
control.get_len()
```

```python
# Generate histograms ==========================
figures = Basic_Plotters(control.cat)
hists_1 = figures.plot_3_hist(fig_nm = '01_clean_gaia_output_1.pdf', x2_bins=7, x3_bins=5, ftsize = 40)
```

## Create Control Sample

```python
# Apply 3rd quality-control ========
cat_0 = control.cat.copy()        # Useful to identify outliers (see below)
control.apply_col_cut(inp_col='parallax', inner=5, outer=9)
control.apply_col_cut(inp_col='pmra',     inner=-20.0, outer=0)
control.get_len()

figures = Basic_Plotters(control.cat)
hists_1 = figures.plot_3_hist(fig_nm = '01_clean_gaia_output_2.pdf', x1_bins=5, x2_bins=3, x3_bins=3, ftsize = 40, x3lim=[-30,-18])
```

```python
# Extract Outliers =================
print(f'Sanity Check: lenght before/after outlier removal:  {len(cat_0)} / {len(control.cat)}')

unc = np.setdiff1d(cat_0['source_id'], control.cat['source_id'])
unc = Table([unc], names=['source_id'])
unc = join(unc, cat_0)
unc = unc['col2mass', 'parallax', 'pmra', 'pmdec']

unc.write('OPH___outliers.tex', format = 'ascii.latex', overwrite = True)
!open OPH___outliers.tex
unc
```

```python
# Sanity Check =================================
len(control.cat), len(unique(control.cat, keys='source_id')), len(unique(control.cat, keys='col2mass')), len(unique(control.cat, keys=['col2mass', 'source_id']))
```

```python
# Look at the radial velocities ================
control.get_vrad()
print()
```

```python
# Get Spatial Extension ========================
print(f"RA ranges from:  {control.cat['ra'].min():10.1F} : {control.cat['ra'].max():5.1F}")
print(f"DEC ranges from: {control.cat['dec'].min():10.1F} : {control.cat['dec'].max():5.1F}")

dist_m = 1000./control.cat['parallax'].mean()
dist_u = dist_m - 1000./(control.cat['parallax'].mean() + control.cat['parallax'].std())
dist_b = 1000./(control.cat['parallax'].mean() - control.cat['parallax'].std()) - dist_m

print(f'Average distance: {dist_m:9.1F} - {dist_u:3.1F} + {dist_b:3.1F}')
print()

# Compute stats ================================
control.get_stats_all()
control.stats_all.write('table_1.tex', format = 'ascii.latex')
control.cat.write('OPH___control_sample.vot', format = 'votable', overwrite = True)
!open table_1.tex
control.stats_all
```
